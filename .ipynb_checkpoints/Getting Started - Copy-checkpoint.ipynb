# 3. Load Data

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
tf.keras.utils.image_dataset_from_directory?
```

```python
data = tf.keras.utils.image_dataset_from_directory('data', class_names=["buildings","forest","glacier","mountain","sea","street"],image_size=(150,150))
test = tf.keras.utils.image_dataset_from_directory('test', class_names=["buildings","forest","glacier","mountain","sea","street"],image_size=(150,150))
```

```python
data
```

```python
data_iterator = data.as_numpy_iterator()
```

```python
batch = data_iterator.next()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

# 4. Scale Data

```python
test = test.map(lambda x,y: (x/255, y))
```

```python
data = data.map(lambda x,y: (x/255, y))
```

```python
data.as_numpy_iterator().next()
```

# 5. Split Data

```python
train_size = int(len(data)*.8)
val_size = int(len(data)*.2)
```

```python
train_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
```

# 6. Build Deep Learning Model

```python
train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(150,150,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
#model.add(Conv2D(16, (3,3), 1, activation='relu'))
#model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(150, activation='relu'))
model.add(Dense(6, activation='softmax'))
```

```python
model.compile?
```

```python
model.compile('adam', loss=tf.losses.sparse_categorical_crossentropy, metrics=['accuracy'])
```

```python
model.summary()
```

# 7. Train

```python
logdir='logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=10, validation_data=val, callbacks=[tensorboard_callback])
```

# 8. Plot Performance

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
test.shape
test_loss, test_acc = model.evaluate(test)
```
